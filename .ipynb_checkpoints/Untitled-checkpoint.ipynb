### Передобробка даних


Перетворення цін в єдину валюту (долари).

```python
import requests

# Функція для отримання курсу валют
def get_exchange_rate(base_currency, target_currency):
    url = f"https://api.exchangerate-api.com/v4/latest/{base_currency}"
    response = requests.get(url)
    data = response.json()
    return data['rates'].get(target_currency)

# Функція для ідентифікації валюти та перетворення цін
def identify_currency(price_str):
    if 'грн' in price_str:
        return 'гривня'
    elif '$' in price_str:
        return 'доллар'
    elif '€' in price_str:
        return 'євро'
    else:
        return None

def convert_to_usd(price_str, currency):
    # Видаляємо пробіли з рядка ціни перед перетворенням у число
    price_str = price_str.replace(' ', '')
    
    if currency == 'гривня':
        exchange_rate = get_exchange_rate('UAH', 'USD')  # Отримуємо актуальний курс гривні до долара
        return float(price_str.replace('грн', '').strip()) * exchange_rate
    elif currency == 'доллар':
        return float(price_str.replace('$', '').strip())
    elif currency == 'євро':
        exchange_rate = get_exchange_rate('EUR', 'USD')  # Отримуємо актуальний курс євро до долара
        return float(price_str.replace('€', '').strip()) * exchange_rate
    else:
        return None

# Примінення функції convert_to_usd до кожного рядка стовпця 'Ціна'
data['Ціна_USD'] = data['Ціна'].apply(lambda price_str: convert_to_usd(price_str, identify_currency(price_str)))
```

Перетворення "Загальна площа" та "Площа кухні" в числовий формат.

```python
# Перетворення "Загальна площа" в числовий формат та переформатування назви стовпця
data['Загальна площа'] = data['Загальна площа'].str.replace(' м²', '').astype(float)
data = data.rename(columns={'Загальна площа': 'Загальна площа (м²)'})

# Перетворення "Площа кухні" в числовий формат та переформатування назви стовпця
data['Площа кухні'] = data['Площа кухні'].str.replace(' м²', '').astype(float)
data = data.rename(columns={'Площа кухні': 'Площа кухні (м²)'})
```

### Перевірка даних на аномалії

Для виявлення аномалій у даних про нерухомість використовувався метод міжквартильного діапазону (IQR). Це дозволяє визначити екстремальні значення, які суттєво відрізняються від інших значень у датасеті.

```python
# Обчислення IQR для площі
Q1 = data_cleaned['Загальна площа (м²)'].quantile(0.25)
Q3 = data_cleaned['Загальна площа (м²)'].quantile(0.75)
IQR = Q3 - Q1

# Визначення меж для аномалій
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Ідентифікація аномалій
anomalies = data_cleaned[(data_cleaned['Загальна площа (м²)'] < lower_bound) | (data_cleaned['Загальна площа (м²)'] > upper_bound)]
print(f"Кількість аномалій: {len(anomalies)}")
print(anomalies)
```

У даному прикладі було виявлено 45 аномалій.
Аномалії мають різні характеристики, такі як ціна, загальна площа, кількість кімнат та інші параметри.
Опис аномалій:
Ціна (USD): Від $25000 до $290000
Загальна площа (м²): Від 121 до 230 м²
Кількість кімнат: Від 1 до 5+ кімнат
Опалення: Переважно "Індивідуальне газове"
Ремонт: Різні типи, але більшість мають "Євроремонт"

Для кращого розуміння розподілу аномалій  створимо графік розподілу цін на нерухомість відносно площі з виділенням аномалій іншим кольором.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Візуалізація розподілу аномалій
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data_cleaned, x='Загальна площа (м²)', y='Ціна_USD')
sns.scatterplot(data=anomalies, x='Загальна площа (м²)', y='Ціна_USD', color='red')
plt.title('Розподіл цін на нерухомість відносно площі')
plt.xlabel('Загальна площа (м²)')
plt.ylabel('Ціна (USD)')
plt.legend(['Дані', 'Аномалії'], loc='upper right')
plt.show()
```

Проведений аналіз дозволив виявити 45 аномалій у даних про нерухомість.Як видно з графіку розсіювання, оголошення з високою ціною відповідають квартирам з великою площею. Аномалій у стовпці не спостерігається.

### Чинники які впливають на ціну

```python
# Підготовка даних для візуалізації - розрахунок сводної таблиці з медіанними значеннями вартості у групах загальної площі
data_pivot = data_cleaned.pivot_table(index='Загальна площа (м²)', values='Ціна_USD', aggfunc='median').reset_index()
data_pivot.columns = ['Загальна площа (м²)', 'Медіанна ціна (USD)']
print(data_pivot)

# Побудова діаграми розсіювання
data_pivot.sort_values('Загальна площа (м²)').plot(x='Загальна площа (м²)', y='Медіанна ціна (USD)', kind='scatter', alpha=0.5, figsize=(10, 6))
plt.title('Залежність ціни від загальної площі для всіх квартир')
plt.xlabel('Загальна площа (м²)')
plt.ylabel('Медіанна ціна (USD)')
plt.grid(True)
plt.show()
```

Коефіцієнт підтверджує сильний позитивний зв'язок між цими змінними.

На діаграмі розсіювання чітко видно лінійну залежність - зі збільшенням загальної площі зростає і ціна. Найбільш чітко ця залежність помітна в діапазоні до 100 кв.м.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Підготовка даних для візуалізації - розрахунок зведеної таблиці з медіанними значеннями вартості у групах площі кухні
data_pivot = data_cleaned.pivot_table(index='Площа кухні (м²)', values='Ціна_USD', aggfunc='median').reset_index()
data_pivot.columns = ['Площа кухні (м²)', 'Медіанна ціна (USD)']
print(data_pivot)

# Побудова діаграми розсіювання
plt.figure(figsize=(10, 6))
plt.scatter(data_pivot['Площа кухні (м²)'], data_pivot['Медіанна ціна (USD)'], alpha=0.5)
plt.title('Залежність ціни від площі кухні для всіх квартир')
plt.xlabel('Площа кухні (м²)')
plt.ylabel('Медіанна ціна (USD)')
plt.grid(True)
plt.show()

# Розрахунок коефіцієнта кореляції Пірсона між площею кухні та ціною
pearson_correlation = data_cleaned['Площа кухні (м²)'].corr(data_cleaned['Ціна_USD'])
print(f"Коефіцієнт кореляції Пірсона між площею кухні та ціною: {pearson_correlation:.2f}")
```

На діаграмі розсіювання чітко видно залежність - зі збільшенням площі кухні зростає ціна.
Коефіцієнт кореляції зі значенням 0,50 також підтверджує наявність залежності вартості площі кухні

```python
# Підготовка даних для візуалізації - розрахунок зведеної таблиці з медіанними значеннями вартості у групах кількості кімнат
data_pivot = data_cleaned.pivot_table(index='Кількість кімнат', values='Ціна_USD', aggfunc='median').reset_index()
data_pivot.columns = ['Кількість кімнат', 'Медіанна ціна (USD)']
print(data_pivot)

# Побудова діаграми розсіювання
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data_pivot, x='Кількість кімнат', y='Медіанна ціна (USD)', alpha=0.5)
plt.title('Залежність ціни від кількості кімнат для всіх квартир')
plt.xlabel('Кількість кімнат')
plt.ylabel('Медіанна ціна (USD)')
plt.grid(True)
plt.show()

# Розрахунок коефіцієнта кореляції Пірсона між кількістю кімнат та ціною
# Спочатку потрібно перетворити кількість кімнат на числовий формат
data_cleaned['Кількість кімнат (число)'] = data_cleaned['Кількість кімнат'].replace({'1 кімната': 1, '2 кімнати': 2, '3 кімнати': 3, '4 кімнати': 4, '5+ кімнат': 5})
pearson_correlation = data_cleaned['Кількість кімнат (число)'].corr(data_cleaned['Ціна_USD'])
print(f"Коефіцієнт кореляції Пірсона між кількістю кімнат та ціною: {pearson_correlation:.2f}")
```

Коефіцієнт кореляції зі значенням 0,50 підтверджує наявність залежності вартості кількості кімнат.
Діаграма розсіювання показує залежність між кількістю кімнат та ціною для всіх квартир. Загалом, зростання кількості кімнат веде до збільшення ціни.

```python
# Підготовка даних для візуалізації - розрахунок зведеної таблиці з медіанними значеннями вартості у групах поверхів
data_pivot = data_cleaned.pivot_table(index='Поверх', values='Ціна_USD', aggfunc='median').reset_index()
data_pivot.columns = ['Поверх', 'Медіанна ціна (USD)']
print(data_pivot)

# Побудова діаграми розсіювання
plt.figure(figsize=(12, 8))
sns.scatterplot(data=data_pivot, x='Поверх', y='Медіанна ціна (USD)', alpha=0.7)
plt.title('Залежність ціни від поверху для всіх квартир')
plt.xlabel('Поверх')
plt.ylabel('Медіанна ціна (USD)')
plt.xticks(ticks=range(int(data_pivot['Поверх'].min()), int(data_pivot['Поверх'].max())+1))
plt.grid(True)
plt.show()
```

Діаграма розсіювання показує залежність між поверхом та ціною для всіх квартир. Загалом, певні поверхи можуть бути більш популярними і, відповідно, мати вищу ціну. Наприклад, поверхи з дробовими значеннями (1.2, 5.6, 9.1) мають значно вищу медіанну ціну, що може вказувати на особливі типи квартир, такі як пентхауси або квартири з особливими характеристиками.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Підготовка даних для візуалізації - розрахунок зведеної таблиці з медіанними значеннями вартості у групах типів будинків
data_pivot = data_cleaned.pivot_table(index='Тип будинку', values='Ціна_USD', aggfunc='median').reset_index()
data_pivot.columns = ['Тип будинку', 'Медіанна ціна (USD)']
print(data_pivot)

# Побудова діаграми стовпців
plt.figure(figsize=(12, 8))
sns.barplot(data=data_pivot, x='Тип будинку', y='Медіанна ціна (USD)')
plt.title('Залежність ціни від типу будинку')
plt.xlabel('Тип будинку')
plt.ylabel('Медіанна ціна (USD)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Перейменування колонок для використання англійських символів
data_cleaned_renamed = data_cleaned.rename(columns={'Тип будинку': 'Building_Type', 'Ціна_USD': 'Price_USD'})

# Підготовка даних для моделі ANOVA
model = ols('Price_USD ~ C(Building_Type)', data=data_cleaned_renamed).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

Використання коефіцієнта кореляції Пірсона не підходить для категоричних змінних, таких як тип будинку.
Ми використали ANOVA (Аналіз дисперсії) для оцінки впливу типу будинку на ціну.Тип будинку має значний вплив на ціну нерухомості, що підтверджується дуже низьким p-значенням (1.062466e-24).
Значення F: Високе значення F (14.695135) свідчить про те, що є значна різниця в середніх цінах між різними типами будинків.

```python
# Підготовка даних для візуалізації - розрахунок зведеної таблиці з медіанними значеннями вартості у групах виду об'єкта
data_pivot = data_cleaned.pivot_table(index='Вид об\'єкта', values='Ціна_USD', aggfunc='median').reset_index()
data_pivot.columns = ['Вид об\'єкта', 'Медіанна ціна (USD)']
print(data_pivot)

# Побудова діаграми стовпців
plt.figure(figsize=(12, 8))
sns.barplot(data=data_pivot, x='Вид об\'єкта', y='Медіанна ціна (USD)')
plt.title('Залежність ціни від виду об\'єкта')
plt.xlabel('Вид об\'єкта')
plt.ylabel('Медіанна ціна (USD)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

Новобудови мають вищу медіанну ціну порівняно з вторинним ринком.
 Новобудови часто мають сучасні зручності, вищу якість будівництва та інші переваги, що підвищують їх вартість. Медіанна ціна 49,900 USD. Вторинні будинки можуть мати старіше обладнання, потребувати ремонтів та бути менш привабливими для покупців.
